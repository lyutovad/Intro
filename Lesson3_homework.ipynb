3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

Окружность

```python
x = []
y1 = []
y2 = []
x0 = -5
y0 = 18
# M = [x0, y0]
r = 25
for i in range(x0-r, x0+r+1):
    x_ = i
    x.append(x_)
    y1 = np.append(y1, [y0 + math.sqrt(r**2 - (x_-x0)**2)])
    y2 = np.append(y2, [y0 - math.sqrt(r**2 - (x_-x0)**2)])
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)
Z = 2*X - 3*Y
Z2 = 2*X - 3*Y - 45
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
